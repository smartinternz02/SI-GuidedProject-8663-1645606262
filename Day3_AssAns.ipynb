```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

# Line Graph

```python
x=np.array([1,2,3,4,5,6,7])
y=np.power(x,2)
plt.plot(x,y)
```

# Scatter Plot

```python
m=np.array([1,2,3,4,5,6,7])
n=np.power(m,3)
plt.scatter(m,n,color='red',marker='*')
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25,5.25],[50,40,20,10,30,20],label="student1",width=0.5)
plt.bar([0.75,1.75,2.75,3.75,4.75,5.75],[30,20,20,40,50,20],label="student2",width=0.5)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.title("Marks of 2 students")
```

# Histogram

```python
a = np.random.normal(170, 10, 250)
plt.hist(a,color='Magenta')
```

# CountPlot

```python
df = pd.DataFrame(['female','female','male','male','female','male','female','female','male','male','female','female','male'],columns=['Gender'])
sns.countplot(x ='Gender', data = df)
plt.show()
```

# HeatMap

```python
data=[{2,3,4},{6,3,5},{6,3,5},{3,7,5},{2,8,1}]
Index= ['I1', 'I2','I3','I4','I5']
Cols = ['C1', 'C2', 'C3']
df = pd.DataFrame(data, index=Index, columns=Cols)
plt.pcolor(df)
plt.show()
```

# BoxPlot

```python
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```

# PairPlot

```python
df = sns.load_dataset('tips')
sns.pairplot(df,hue ='size')
plt.show()
```

# JoinPlot

```python
data = sns.load_dataset("attention")
sns.jointplot(x = "solutions", y = "score",kind = "reg", data = data)
plt.show()
```

```python
data = sns.load_dataset("attention")
sns.jointplot(x = "solutions", y = "score",kind = "hex", data = data)
plt.show()
```
